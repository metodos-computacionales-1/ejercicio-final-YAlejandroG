```python
import matplotlib.pylab as plt
import numpy as np
from IPython.display import clear_output
```

```python
%%bash
g++ Pendulo.cpp
./a.out 10000000 >Pendulo.txt
```

```python
Pendulo = np.genfromtxt("Pendulo.txt")

time = Pendulo[:,0]
theta = Pendulo[:,1]
omega = Pendulo[:,2]
```

```python
plt.figure(figsize=(15,4))
plt.plot(time,theta,c='g')
plt.grid()
plt.xlabel('time (s)')
plt.ylabel('$\Theta$ (rads)')
plt.title('Fd = 1.4 N')
plt.show()
```

```python
for i in range(10,len(theta),100):
    plt.scatter(theta[:i],omega[:i],s=5,c='g',label='t = '+str(time[i])+'s')
    plt.ylim(-3,1.5)
    plt.xlim(-3.5,3.5)
    plt.grid()
    plt.legend()
    plt.xlabel('$\Theta$ (rads)')
    plt.ylabel('$\omega$ (rads)')
    plt.savefig(str(i).zfill(4)+'.png')
    plt.close()
    if i>9910:
        break;
```
